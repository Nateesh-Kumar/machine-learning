```python
# @title Default title text
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample dataset
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DATASET/bike-sharing.csv")

# Select the actual numerical data for clustering
# Use the column names directly instead of creating a new DataFrame
x_data = data['cnt']
y_data = data['registered']

data_for_kmeans = pd.DataFrame({
    'x': x_data,
    'y': y_data
})

# Elbow method to determine the optimal number of clusters
wcss = []
for i in range(1, data.shape[0] + 1):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(data_for_kmeans)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow graph
plt.plot(range(1, data.shape[0] + 1), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Applying K-means with the chosen number of clusters, e.g., K=2
kmeans = KMeans(n_clusters=i, random_state=42)
kmeans.fit(data_for_kmeans)

# Get the cluster centroids and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Add the cluster labels to the dataframe
data_for_kmeans['Cluster'] = labels

# Plot the clustered data_for_kmeans
plt.scatter(data_for_kmeans['x'], data_for_kmeans['y'], c=data_for_kmeans['Cluster'], cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='x')  # Mark centroids
plt.title('K-means Clustering')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 2: Generate synthetic data
np.random.seed(42)
X = np.vstack([
    np.random.normal(loc=0.0, scale=0.5, size=(50, 2)),
    np.random.normal(loc=3.0, scale=0.5, size=(50, 2)),
    np.random.normal(loc=6.0, scale=0.5, size=(50, 2))
])

# Step 3: Apply K-Means clustering
kmeans_model = KMeans(n_clusters=3, random_state=42)
kmeans_model.fit(X)

# Step 4: Visualize the results
plt.scatter(X[:, 0], X[:, 1], c=kmeans_model.labels_, cmap='rainbow', label="points")
plt.title('K-Means Clustering Example')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
