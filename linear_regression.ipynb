**variables**

```python
m = 0
c = 1
n = len(x)
learning_rate = 0.0001
iterations = 1000
```

```python

for i in range(iterations):
  y_pred = m * x + c

    # Loss function (Mean Squared Error)
  loss = np.sum((y_pred - y) ** 2) / n

    # slope and intercept updating
  gradient_m = (2 / n) * np.sum((y_pred - y) * x)
  gradient_c = (2 / n) * np.sum(y_pred - y)
  m = m - learning_rate * gradient_m
  c = c - learning_rate * gradient_c
```

```python
plt.scatter(x, y, color='blue', label='Data')
y_best_fit = m * x + c
plt.plot(x, y_best_fit, color='red', label='Best-fit line')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x=np.array(x).reshape(-1, 1)  # Feature
y = np.array(y)
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
slope = model.coef_[0]
intercept = model.intercept_

print("Slope:", slope)
print("Intercept:", intercept)
```
